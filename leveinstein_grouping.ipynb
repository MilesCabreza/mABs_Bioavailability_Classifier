```python
""" Import packages
"""
# Utility
import numpy as np
import pandas as pd
from scipy import stats as st

!pip install levenshtein
from Levenshtein import distance
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import pdist
```

```python
#AZ data
df = pd.read_csv("Bioavailability_sequences.csv", delimiter = ",")
df
```

```python
## pairwise Levenshtein distance
distance_seq= np.zeros([len(df), len(df)])
for i, r1 in df.iterrows():
    for j, r2 in df.iterrows():
        if i>j:
            mabs_seq1= r1["Light"]+r1["Heavy"]
            mabs_seq2= r2["Light"]+r2["Heavy"]
            distance_seq[i, j]=distance(mabs_seq1, mabs_seq2)
            distance_seq[j, i]= distance_seq[i, j]
```

```python
# Perform hierarchical clustering
Z = linkage(pdist(distance_seq), 'complete')

# Define the threshold for cutting the dendrogram
threshold =75

# Assign cluster labels to each data point
clusters = fcluster(Z, t=threshold, criterion='distance')


x=np.max(clusters)
```

```python
for i in range(x):
    if np.sum(np.array(clusters==i, dtype=int))>0:
        print(i)
        print([list(df.mAbName)[x] for x in list(np.where(clusters==i)[0])])
```

```python
df['clusters'] = clusters
df
```
